```python
def max_consecutive_elements(input_str):
    result, cur_letter_idx = 0, 0
    while cur_letter_idx < len(input_str):
        next_letter_idx = cur_letter_idx
        while next_letter_idx < len(input_str) \
        and input_str[next_letter_idx] == input_str[cur_letter_idx]:
            next_letter_idx += 1
        result = max(result, next_letter_idx - cur_letter_idx)
        cur_letter_idx = next_letter_idx
    return result

max_consecutive_elements("ЯЯблоки в садууу")
```

```python
# Метод двух указаталей работает в монотонных функциях если они возрастают или убывают (строго возрастают или строго убывают)

arr = [] # Строго возрастающий массив
target = 1


def max_arr_string(arr, target):
    left, right, current_sum = 0, 0, 0
    while left != (len(arr) - 2) and right != len(arr) - 1 and current_sum != target:
        while current_sum < target and current_sum != target and right < len(arr) - 1:
            right += 1
            current_sum = sum(arr[left:right+1])
        while current_sum > target and current_sum != target and left < len(arr) - 2:
            left += 1
            current_sum = sum(arr[left:right+1])
        if current_sum == target:
            return True
    else:
        return False
    

max_arr_string(arr, target)
```

```python
def subarray_sum(non_negative_arr, target):
    right, current_sum = 0, 0
    for left in range(len(non_negative_arr)):
        # пересчитываем сумму
        if left > 0:
            current_sum -= non_negative_arr[left - 1]

        # при необходимости двигаем правую границу
        while right < len(non_negative_arr) and current_sum < target:
            current_sum += non_negative_arr[right]
            right += 1

        if current_sum == target:
            return True

    return False 

arr = [0, 1, 2, 3, 4, 5, 6] # Строго возрастающий массив
target = 0

subarray_sum(arr, target)
```

```python
# Бинарный поиск

def binary_search(arr, target):
    left, right = 0, len(arr)
    while left < right:
        mid = (left + right) // 2
        if target == arr[mid]:
            return True
        elif target > arr[mid]:
            left = mid + 1
        else:
            right = mid
    else:
        return False
            
binary_search([q for q in range(0, 101)], 2)
```

```python
def solve_equation(value):
    left, right = 1, value
    for _ in range(100):
        mid = (left + right) / 2
        expr_result = mid * log2(mid)
        if expr_result < value:
            left = mid
        else:
            right = mid
    return mid

solve_equation(8)
```

```python
def max_lower_or_equal(sorted_arr, value):
    # Сначала проверим, существует ли искомый элемент
    if not sorted_arr or sorted_arr[0] > value:
        return -1

    left_idx, right_idx = 0, len(sorted_arr)
    while left_idx + 1 < right_idx:
        mid_idx = (left_idx + right_idx) // 2
        if sorted_arr[mid_idx] <= value:
            left_idx = mid_idx
        else:
            right_idx = mid_idx
    return left_idx

max_lower_or_equal([q for q in range(1, 2)], 0)
```
